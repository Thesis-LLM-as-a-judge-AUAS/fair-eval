# Vanilla GPT

```python
GPT_QUESTION_NUMBER = 1
GPT_VERSION = 'gpt4'
```

```python
vanilla_questions_df = pd.read_json(f'./sampled/questions/{GPT_QUESTION_NUMBER}.jsonl', lines=True)
vanilla_questions_df
```

```python
vanilla_gpt35_df = pd.read_json(f'./sampled/answer_gpt35/{GPT_QUESTION_NUMBER}.jsonl', lines=True)
vanilla_gpt35_df
```

```python
vanilla_vicuna_df = pd.read_json(f'./sampled/answer_vicuna/{GPT_QUESTION_NUMBER}.jsonl', lines=True)
vanilla_vicuna_df
```

```python
vanilla_merged_df = vanilla_vicuna_df \
    .merge(vanilla_gpt35_df, on='question_id', how='right', suffixes=('_vicuna', '_gpt')) \
    .merge(vanilla_questions_df, on='question_id', how='right', suffixes=('', '_question'))

vanilla_merged_df
```

```python
vanilla_dropped_df = vanilla_merged_df.drop(
    ['metadata_vicuna', 'metadata_gpt', 'model_id_vicuna', 'model_id_gpt', 'answer_id_gpt', 'answer_id_vicuna'],
    axis=1).rename(columns={'text': 'question_text'})

vanilla_dropped_df
```

```python
vanilla_dropped_df.to_json(f'./vicuna/vanilla-vicuna-{GPT_VERSION}.json', orient='records', lines=False, force_ascii=False)
```

# MEC + BPC with GPT-3.5

```python
MEC_QUESTION_NUMBER = 3
```

```python
mec_questions_df = pd.read_json(f'./sampled/questions/{MEC_QUESTION_NUMBER}.jsonl', lines=True)
mec_questions_df
```

```python
mec_gpt35_df = pd.read_json(f'./sampled/answer_gpt35/{MEC_QUESTION_NUMBER}.jsonl', lines=True)
mec_gpt35_df
```

```python
mec_vicuna_df = pd.read_json(f'./sampled/answer_vicuna/{MEC_QUESTION_NUMBER}.jsonl', lines=True)
mec_vicuna_df
```

```python
mec_merged_df = mec_vicuna_df \
    .merge(mec_gpt35_df, on='question_id', how='right', suffixes=('_vicuna', '_gpt')) \
    .merge(mec_questions_df, on='question_id', how='right', suffixes=('', '_question'))

mec_vicuna_df
```

```python
mec_dropped_df = mec_merged_df.drop(
    ['metadata_vicuna', 'metadata_gpt', 'model_id_vicuna', 'model_id_gpt', 'answer_id_gpt', 'answer_id_vicuna'],
    axis=1).rename(columns={'text': 'question_text'})

mec_dropped_df
```

```python
mec_expanded = mec_dropped_df.loc[mec_dropped_df.index.repeat(6)].reset_index(drop=True)

mec_expanded['repeat_id'] = np.tile(np.arange(1, 7), len(mec_dropped_df))

mec_expanded['is_reversed'] = mec_expanded['repeat_id'] > 3

mask = mec_expanded['is_reversed']
swapped = mec_expanded.loc[mask, ['text_vicuna', 'text_gpt']].copy()
mec_expanded.loc[mask, 'text_vicuna'] = swapped['text_gpt']
mec_expanded.loc[mask, 'text_gpt'] = swapped['text_vicuna']

mec_expanded.drop(columns=['repeat_id'], inplace=True)
mec_expanded
```

```python
mec_expanded.to_json(f'./vicuna/mec-bpc-vicuna-{GPT_VERSION}.json', orient='records', lines=False, force_ascii=False)
```

# Answers extraction with MEC and BPC

```python
MEC_PRESET_NAME = 'review_gpt35_vicuna_gpt-4_mec3_bpc1'
```

```python
def parse_score_from_review(review):
    first_score = review.split("\n")[-2].split(":")[-1].strip()
    second_score = review.split("\n")[-1].split(":")[-1].strip()
    return [float(first_score), float(second_score)]
```

```python
def extract_sub_scores(score_path):
    scores = pd.read_json(score_path, lines=True)
    scores_expanded = scores.loc[scores.index.repeat(6)].reset_index(drop=True)

    for index, row in scores_expanded.iterrows():
        review_feature = "review" if index % 6 < 3 else "review_bpc"
        row_review = row[review_feature][index % 3]

        scores_expanded.at[index, 'score'] = parse_score_from_review(row_review)
        
    return scores_expanded
```

```python
for i in range(1, 47):
    result_scores = extract_sub_scores(f'./sampled_review/{i}/{MEC_PRESET_NAME}.jsonl')
    result_scores.to_json(f'./expanded_review/{MEC_PRESET_NAME}/{i}_{MEC_PRESET_NAME}.jsonl', orient='records', lines=True, force_ascii=False)
```
