```python
import pandas as pd
```

# Vanilla GPT-3.5

```python
vanilla_questions_df = pd.read_json('./sampled/questions/0.jsonl', lines=True)
vanilla_questions_df
```

```python
vanilla_gpt35_df = pd.read_json('./sampled/answer_gpt35/0.jsonl', lines=True)
vanilla_gpt35_df
```

```python
vanilla_vicuna_df = pd.read_json('./sampled/answer_vicuna/0.jsonl', lines=True)
vanilla_vicuna_df
```

```python
vanilla_merged_df = vanilla_vicuna_df \
    .merge(vanilla_gpt35_df, on='question_id', how='right', suffixes=('_vicuna', '_gpt')) \
    .merge(vanilla_questions_df, on='question_id', how='right', suffixes=('', '_question')).set_index('question_id')

vanilla_merged_df
```

```python
vanilla_dropped_df = vanilla_merged_df.drop(
    ['metadata_vicuna', 'metadata_gpt', 'model_id_vicuna', 'model_id_gpt', 'answer_id_gpt', 'answer_id_vicuna'],
    axis=1).rename(columns={'text': 'question_text'})

vanilla_dropped_df
```

```python
vanilla_dropped_df.to_json('./vicuna/vanilla-vicuna.json', orient='records', lines=False, force_ascii=False)
```
